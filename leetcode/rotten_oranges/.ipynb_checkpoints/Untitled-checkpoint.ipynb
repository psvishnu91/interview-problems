```python
import numpy as np
from scipy import optimize
from mpl_toolkits import mplot3d
```

```python
def f(x, y):
    return np.exp(-(2*x*x + y*y - x*y) / 2)


def fvec(X):
    return 
```

```python
x = np.arange(-2,2, 0.1)
y = np.arange(-2,2, 0.1)
xv, yv = np.meshgrid(x, y)
```

```python
fxy = f(xv, yv)
```

```python
def dfdx (x, y) :
    return 1/2 * (-4*x + y) * f(x, y)

def dfdy (x, y) :
    return 1/2 * (x - 2*y) * f(x, y)


def gradf(xy):
    x, y = xy
    return np.array(
        [
            dfdx(x, y),
            dfdy(x, y),
        ],
    )
```

```python
x0, y0 = -1, -1
xsol, ysol = optimize.root(gradf, [x0, y0]).x
```

```python
print(f"{xsol=:.1f}, {ysol=:.1f}")
print(f"{gradf([xsol, ysol])=}")
print(f"{f(xsol, ysol)=:.1f}")
```
